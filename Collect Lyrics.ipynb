```python
import datetime
import pylyrics3 as pl
from billboard import ChartData
from functools import reduce
from multiprocessing.pool import ThreadPool, Pool
```

```python
def from_time(time_str):
    '''Generator for dates from a given "YYYY-MM-DD" string until now'''
    now = datetime.datetime.now()
    week_delta = datetime.timedelta(days=7)
    yyyymmdd = map(lambda x: int(x), time_str.split('-'))
    start_date = datetime.datetime(*yyyymmdd)
    while start_date < now:
        yield start_date
        start_date += week_delta
```

```python
dates = from_time('2015-01-01')
with ThreadPool(4) as pool:
    charts = pool.map(lambda date: ChartData('hot-100', date.strftime('%Y-%m-%d')), dates, chunksize=4)
    artists = map(lambda chart: map(lambda entry: entry.artist.lower().strip(), chart), charts)
    artists_set = reduce(lambda prev, curr: prev.union(curr), artists, set())
    
```

```python
artists_set_no_feat = set(map(lambda x: x.split('featuring')[0], artists_set))
artist_lyrics = map(lambda artist: {artist: pl.get_artist_lyrics(artist)}, artists_set_no_feat)
artist_lyrics_dict = reduce(lambda prev, curr: {**prev, **curr}, artist_lyrics, dict())
```

```python
import json
with open('lyrics.json', 'w') as f:
    json.dump(artist_lyrics_dict, f)
```
