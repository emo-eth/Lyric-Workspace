```python
def from_time(time_str, until=None):
    '''Generator for dates from a given "YYYY-MM-DD" string until now'''
    if until is None:
        until = datetime.datetime.now()
    else:
        yyyymmdd = map(lambda x: int(x), until.split('-'))
        until = datetime.datetime(*yyyymmdd)
    week_delta = datetime.timedelta(days=7)
    yyyymmdd = map(lambda x: int(x), time_str.split('-'))
    start_date = datetime.datetime(*yyyymmdd)
    while start_date < until:
        yield start_date
        start_date += week_delta
```

```python
dates = from_time('1970-01-01', '1980-01-01')
with ThreadPool(4) as pool:
    charts = pool.map(lambda date: ChartData('hot-100', date.strftime('%Y-%m-%d'), timeout=5), dates, chunksize=4)
    artists = map(lambda chart: map(lambda entry: entry.artist.lower().strip(), chart), charts)
    artists_set = reduce(lambda prev, curr: prev.union(curr), artists, set())
    
```

```python
for chart in charts:
    with open('./charts/' + chart.date + '.json', 'w') as f:
        d = json.loads(chart.json())
        json.dump(d, f)
```

```python
artists_set = set()
for fname in glob.glob('./charts/*.json'):
    with open(fname) as f:
        chart = json.load(f)
        artists = set(map(lambda x: x['artist'], chart['entries']))
        artists_set.update(artists)
```

```python
artists_set = map(lambda artist: artist.lower().strip(), artists_set)
artists_set = set(map(lambda x: x.split('featuring')[0], artists_set))
len(artists_set)
```

```python
with open('lyrics.json') as f:
    lyrics_json = json.load(f)
known_artists = set(lyrics_json.keys())
for fname in glob.glob('./lyrics/*.json'):
    with open(fname) as f:
        lyrics = json.load(f)
        artist = set(lyrics.keys())
        known_artists.update(artist)
artists_set = map(lambda x: x.replace('/', ' '), artists_set)
artists_set = set(filter(lambda x: x not in known_artists, artists_set))
len(artists_set)
```

```python
def get_lyrics_and_save_json(artist):
    print(artist)
    artist_lyrics = {artist: pl.get_artist_lyrics(artist)}
    with open('./lyrics/' + artist + '.json', 'w') as f:
       json.dump(artist_lyrics, f)
    return artist_lyrics
```

```python
artist_lyrics = map(get_lyrics_and_save_json, artists_set)
artist_lyrics_dict = reduce(lambda prev, curr: {**prev, **curr}, artist_lyrics, dict())
```

```python
artist_lyrics_dict.update(lyrics_json)
with open('lyrics2.json', 'w') as f:
    json.dump(artist_lyrics_dict, f)
```
